```python
import pandas as pd
import numpy as np
import scipy
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn import preprocessing
from sklearn.metrics import classification_report
```

```python
draftwithcollege = pd.read_csv('CollegeRookieStats_TeamRanksAdd.csv')
stats = pd.read_csv('CollegeRookieStats.csv')
standard = pd.read_csv('CollegeRookieStats_standardized.csv')
```

```python
stats.isnull().sum()
```

```python
stats['Pos'] = stats['Pos'].map( {'PF': 0, 'SF': 1, 'C': 2, 'PG': 3, 'SG': 4, 'C-': 2} ).astype(int)
```

```python
standard.isnull().sum()
```

```python
stats.head()
```

```python
d = {range(1, 11): 1, range(11, 21): 2, range(21, 31): 3,range(31, 41): 4,range(41, 51): 5,
     range(51, 61): 6}

standard['DraftRange'] = standard['DraftPick'].apply(lambda x: next((v for k, v in d.items() if x in k), 0))
```

For this, value 1 for DraftRange is Top 10, 2 is Top 20 and 3 is the Top 30, 4 is Top 40 and so on. These are the categories that will be used in the classification model.

```python
standard.tail(30)
```

```python
print(np.unique(standard['DraftRange'], return_counts=True))
print(np.unique(standard['DraftPick'], return_counts=True))
```

Next step will be dropping the values that will not be used in predicting the draft.

```python
x = standard.drop(['ID','Name', 'NBATeam', 'DraftPick', 'Unnamed: 0','NBATRB','NBAAST','NBASTL','NBABLK','NBAPTS','College','DraftYear','DraftRange'], axis = 1)
y = standard['DraftRange']

x_1 = standard[['PTS','MP','FGP','SOS']]
y_1 = standard['DraftRange']
```

I also attempted to use just 4 categories to predict the outcome shown in x_1, for the rest of notebook the models using just the four stats (Points, Minutes Played, Field Goal Percentage and Strength of Schedule) will have a 1 at the end.

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2,stratify = y)
x_train1, x_test1, y_train1, y_test1 = train_test_split(x_1, y_1, test_size = 0.2,stratify = y_1)
```

```python
print(np.unique(y_train, return_counts=True))
print(np.unique(y_test, return_counts=True))
```

## K Nearest Neighbors

```python
from sklearn.neighbors import KNeighborsClassifier
model_KNN = KNeighborsClassifier(n_neighbors=7)
model_KNN.fit(x_train, y_train)
y_testKNN =model_KNN.predict(x_test)
print(classification_report(y_test, y_testKNN, labels=[1, 2, 3, 4, 5, 6]))
```

Now only using the four stats to predict draft range

```python
from sklearn.neighbors import KNeighborsClassifier
model_KNN1 = KNeighborsClassifier(n_neighbors=12)
model_KNN1.fit(x_train1, y_train1)
y_testKNN1 =model_KNN1.predict(x_test1)
print(classification_report(y_test, y_testKNN1, labels=[1, 2, 3, 4, 5, 6]))
```

```python
accuracyKNN = np.mean(np.where(y_testKNN==y_test,1,0))
print(accuracyKNN)

accuracyKNN1 = np.mean(np.where(y_testKNN1==y_test1,1,0))
print(accuracyKNN1)
```

## Random Forest

```python
model_RF = RandomForestClassifier(n_estimators = 400, max_depth = 40)
_ = model_RF.fit(x_train,y_train)

y_testRF = model_RF.predict(x_test)
print(classification_report(y_test, y_testRF, labels=[1, 2, 3, 4, 5, 6]))
```

Now the Random Forest model using only 4 stats to predict draft range.

```python
model_RF1 = RandomForestClassifier(n_estimators = 400, max_depth = 40)
_ = model_RF1.fit(x_train1,y_train1)

y_testRF1 = model_RF1.predict(x_test1)
print(classification_report(y_test, y_testRF1, labels=[1, 2, 3, 4, 5, 6]))
```

```python
accuracyRF = np.mean(np.where(y_testRF==y_test,1,0))
print(accuracyRF)

accuracyRF1 = np.mean(np.where(y_testRF1==y_test1,1,0))
print(accuracyRF1)
```

## Decision Tree

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
model_DT = DecisionTreeClassifier(max_depth=4)
_ = model_DT.fit(x_train, y_train)

y_testDT = model_DT.predict(x_test)
print(classification_report(y_test, y_testDT, labels=[1, 2, 3, 4, 5, 6]))
```

Just the 4 stats to predict draft range

```python
model_DT1 = DecisionTreeClassifier(max_depth=7)
_ = model_DT1.fit(x_train1, y_train1)

y_testDT1 = model_DT1.predict(x_test1)
print(classification_report(y_test, y_testDT1, labels=[1, 2, 3, 4, 5, 6]))
```

```python
accuracyDT = np.mean(np.where(y_testDT==y_test,1,0))
print(accuracyDT)

accuracyDT1 = np.mean(np.where(y_testDT1==y_test1,1,0))
print(accuracyDT1)
```
