## Part 3: Working with Categorical Variables

We will be creating lists to specify the order for each of the three categorical variables of cut, color, and clarity.

```python
clarity_levels = ['I1', 'SI2', 'SI1', 'VS2', 'VS1', 'VVS2', 'VVS1', 'IF']
cut_levels = ['Fair', 'Good', 'Very Good', 'Premium', 'Ideal']
color_levels = ['J', 'I', 'H', 'G', 'F', 'E', 'D']
```

We will now use these lists to communicate to Pandas the correct order for the leves of the three categorical variables

```python
diamonds.clarity = pd.Categorical(diamonds.clarity, clarity_levels)
diamonds.cut = pd.Categorical(diamonds.cut, cut_levels)
diamonds.color = pd.Categorical(diamonds.color, color_levels)
```

We will now create a list of named colors to be used as palettes to be used for later visualizations.

```python
clarity_pal = ['salmon', 'gold', 'limegreen', 'lightseagreen', 'dodgerblue', 'slateblue', 'purple', 'crimson' ]
cut_pal = ['lightcoral', 'orange', 'deepskyblue', 'mediumaquamarine', 'limegreen']
color_pal = ['lightcoral', 'gold', 'limegreen', 'paleturquoise', 'skyblue', 'slateblue', 'darkorchid']
```

## Part 5: Scatterplots of Price Against Carat

Here we will examine the relationship between price and carat by generating a scatter plot.

```python
plt.figure(figsize = [8,6])

plt.scatter(x = diamonds.loc[:,'carat'], y = diamonds.loc[:,'price'],
            s = 20, alpha = 0.4, color = 'salmon')

plt.xlabel('Carat Size')
plt.ylabel('Price (In $)')
plt.title('Relationship between Diamond Price and Carat Size')

plt.show()
```

We will now reproduce the above plot, but color the points according to their clarity.

```python
plt.figure(figsize = [8,6])

for i in range(0, len(clarity_levels)):
     cc = clarity_levels[i]
    
     plt.scatter(x = diamonds.loc[(diamonds.clarity == cc),'carat'], y = diamonds.loc[(diamonds.clarity == cc),'price'],
     s = 20, alpha = 0.4, c = clarity_pal[i], label = cc)

plt.xlabel('Carat Size')
plt.ylabel('Price (In $)')
plt.title('Relationship between Diamond Price and Carat Size')
plt.legend()
plt.show()
```

Belwo we will separate the points into their own plots according to their clarity level.

```python
plt.figure(figsize = [12,6])

for i in range(0, len(clarity_levels)):
    plt.subplot(2,4,1+i)
    cc = clarity_levels[i]
    
    plt.scatter(x = diamonds.loc[(diamonds.clarity == cc),'carat'], y = diamonds.loc[(diamonds.clarity == cc),'price'],
    s = 20, alpha = 0.4, c = clarity_pal[i], label = cc)
    plt.xlim([0,4])
    plt.ylim([0,20000])
    plt.xlabel('Carat Size')
    plt.ylabel('Price (In $)')
    plt.title(cc)


plt.tight_layout()
plt.show()
```

## Part 6: Applying Logarithmic Transformations

In this part, we will start by adding two new columns that will contain the transformed variables for carat and price.

```python
diamonds.loc[:,'ln_carat'] = np.log(diamonds.carat)
diamonds.loc[:,'ln_price'] = np.log(diamonds.price)
```

Below we will use histograms to explore the distribution of the diamond prices, and the log of the diamond prices.

```python
plt.figure(figsize = [12,4])

plt.subplot(1,2,1)

plt.hist(diamonds.loc[:,'carat'], bins = 20,
         edgecolor = 'k', color = 'salmon')
plt.xlabel('Carat Size')
plt.ylabel('Count')
plt.title('Histogram of Carat Sizes')

plt.subplot(1,2,2)

plt.hist(diamonds.loc[:,'ln_carat'], bins = 20,
         edgecolor = 'k', color = 'powderblue')
plt.xlabel('Natural Log of Carat Size')
plt.ylabel('Count')
plt.title('Histogram of Log-Carat Sizes')

plt.show()
```

## Part 7: Scatterplot of Transformed Variables

Below we will create a scatterplot of ln_price against ln_carat, with the points being colored in accordance with their clarity.

```python
plt.figure(figsize = [8,6])

for i in range(0, len(clarity_levels)):
     cc = clarity_levels[i]
    
     plt.scatter(x = diamonds.loc[(diamonds.clarity == cc),'ln_carat'], y = diamonds.loc[(diamonds.clarity == cc),'ln_price'],
     s = 20, alpha = 0.4, c = clarity_pal[i], label = cc)

plt.xlabel('Natural Log of Carat Size')
plt.ylabel('Natural Log of Price')
plt.title('Relationship between Log-Price and Log-Carat Size')
plt.legend()
plt.show()
```

## Part 8: Grouping by Cut

Below, we will creatie a grouped DataFrame displaying the mean price and carat size for each cut level.

```python
gb_cut = diamonds.loc[:,['cut', 'price', 'carat']
                     ].groupby('cut').mean()

gb_cut
```

We will now use bar charts to display the above grouped data.

```python
plt.figure(figsize = [12,4])

plt.subplot(1,2,1)

for i in range(0,len(cut_pal)):
    
    plt.bar(x = gb_cut.index[i], height = gb_cut.iloc[i,0],
            edgecolor = 'k', color = cut_pal[i])
plt.xlabel('Cut')
plt.ylabel('Mean Price')
plt.title('Mean Price by Cut')

plt.subplot(1,2,2)

for i in range(0,len(cut_pal)):
    
    plt.bar(x = gb_cut.index[i], height = gb_cut.iloc[i,1],
            edgecolor = 'k', color = cut_pal[i])

plt.xlabel('Cut')
plt.ylabel('Mean Carat Size by Cut')
plt.title('Mean Carat Size by Cut')


plt.show()
```
